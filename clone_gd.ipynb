Σε περίπτωση που έχετε ήδη το repository κατεβασμένο, μπορείτε να τρέξετε την παρακάτω εντολή για να μεταφερθείτε στον φάκελο του project και στην συνέχεια να κάνετε ενημέρωση των αλλαγών που υπάρχουν στο αποθετήριο. Με τον παρακάτω κώδικα ότι αλλαγές έχετε κάνει στο αποθετήριο σας θα διαγραφούν ώστε να γίνει λήψη του ενημερωμένου περιεχομένου από το git repository. 


````python
# έλεγχος αν το notebook τρέχει στο google colab
try:
    from google.colab import drive
    drive.mount('/content/drive')
    %cd /content/drive/MyDrive/Colab\ Notebooks/gee
    !git reset --hard HEAD # επαναφορά στην τελευταία έκδοση, σε περίπτωση που υπάρχουν τοπικές αλλαγές
    !git pull # ενημέρωση του αποθετηρίου με την τελευταία έκδοση
except:
  print("Παρακαλώ εκτελέστε το notebook στο Google Colab")
````
